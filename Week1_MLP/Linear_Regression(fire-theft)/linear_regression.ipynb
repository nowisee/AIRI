""" simple linear regression """

* import

```python
import tensorflow as tf
from fire_theft import *
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
xy_data = np.array(fire_theft_data)
x_data = xy_data[:, 0]
y_data = xy_data[:, 1]

learning_rate = 0.001
iteration = 500
trained_weight = 0
trained_bias = 0

input_value = 15.5
```

* Placeholder

```python
X = tf.placeholder(tf.float32, name='x')
Y = tf.placeholder(tf.float32, name='y')
```

* Variable

```python
W = tf.Variable(0.0, name='w')
b = tf.Variable(0.0, name='b')
```

* Linear function

```python
Y_pred = W * X + b
```

* Loss function

```python
loss = tf.square(Y_pred - Y)
```

* Optimizer

```python
trainer = tf.train.GradientDescentOptimizer(learning_rate)
updateModel = trainer.minimize(loss)
```

* Training

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(iteration):
        total_loss = 0.0
        for x, y in fire_theft_data:
            _, err = sess.run([updateModel, loss], feed_dict={X:x, Y:y})
            total_loss += err
        print('loss: ', total_loss)
    trained_weight, trained_bias = sess.run([W, b])
```

```python
print('trained value: w = ',trained_weight , ' b = ', trained_bias)
```

```python
y_pred = trained_weight * input_value + trained_bias
print('input : ', input_value, 'predicted value : ', y_pred)
```

```python
plt.plot(x_data, y_data, 'bo', label='Real data')
plt.plot(x_data, trained_weight * x_data + trained_bias, 'r', label='Predicted data')
plt.legend()
plt.show()
```
